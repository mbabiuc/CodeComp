### Spectrogram Plot

```python
Run_List = [
    './THC_Catalog/Strain/Translated/R01\THC0020_LS220_1365_1250_Trim_Freq_Time.txt',
    './BAM_Catalog/Strain/Translated/R01\BAM0005_ALF2_1375_1375_r1000_Trim_Freq_Time.txt',
    './BAM_Catalog/Strain/Translated/R01\BAM0011_ALF2_1500_1500_r800_Trim_Freq_Time.txt',
    './BAM_Catalog/Strain/Translated/R01\BAM0026_G2_1515_1515_r400_Trim_Freq_Time.txt'
]

n1_list = [275, 275, 275, 275]
n2_list = [-275,-1250, -900, -900]

nfft_list = [512, 256, 256, 256]

Id_f2 = [True, True, False, False]

freqs_dis = [True, True, False, True]
```

```python
id_list2 = ['LS220_091_261', 'ALF2_100_275', 'ALF2_100_300', 'G2_100_302']
```

```python
#Create Figure of the 4 Spectrogram plots
#plt.figure(figsize = (10,6))
fig,axes = plt.subplots(2,2, figsize=(10,5))

for k in range(len(Run_List)):
    df = pd.read_csv(Run_List[k], sep='\t')
    
    #Convert to seconds: SACRA time is given in solar masses, and so we can convert as follows
    time_s = (4.925490947641266978197229498498379006e-6)*df['t_translated']

    #print(time_s)
    
    #Compute avg timestep
    avg_dt=sum([time_s[i+1]-time_s[i] for i in range(len(time_s)-1)])/len([time_s[i+1]-time_s[i] for i in range(len(time_s)-1)])
    print(avg_dt)
    print(sum([time_s[i+1]-time_s[i] for i in range(len(time_s)-1)]))
    print(len(time_s))
    
    
    # Slice the data
    data = df['Reh/M'][:]
    time = time_s[:]

    # Define parameters for the spectrogram
    nfft = nfft_list[k]  # Length of each segment
    fs = 1/avg_dt   # Sampling frequency (1/timestep to get samples per second)
    noverlap = 98*nfft // 100  # Overlap between segments

    # Create the spectrogram
    plt.figure(figsize=(10, 6))
    Pxx, freqs, bins, im = plt.specgram(data, NFFT=nfft, Fs=fs, noverlap=noverlap, pad_to = 10000, cmap='viridis') #e.g. inferno, viridis, plasma
    
    #Compute log scale
    logh = np.log10(Pxx)


    #Find dominant frequency over 2nd half of the data where its stabilized in the postmerger
    avg = [0 for i in range(len(logh))]
    for i in range(int(len(logh))):
        #print(i)
        avg[i] = sum(logh[i][int(len(logh[i])/2):]) #Summing up logh over time for each frequency

    f2_index = np.argmax(avg) #finding dominant freq as one with highest total (highest avg) logh
    f2 = freqs[f2_index]

    #print f2 and f2_index
    print('Index:',f2_index, ', Frequency:', f2, 'Hz')
    #print(freqs[30])

    #Create Spectrogram
    df = df
    #plt.figure(figsize=(10, 6))
    #Choose row and column number by k index (0,1,2,3)
    if k%2 == 0:
        col = 0
    else:
        col = 1
    row1 = k//2
    
    f1 = axes[row1][col].pcolormesh(bins,freqs/1000, np.log10(Pxx))
    axes[row1][col].set_ylim(0,8)
    #.ylim(0,8000)

    #Plotting the instantaneous frequency for comparison
    st = n1_list[k]
    end = n2_list[k]
    axes[row1][col].plot(df['t_translated'][st:end]*(4.923e-6), df['frequency'][st:end]/(2*np.pi*4.923e-3), linestyle='dotted', color='r')

    #Find dominant frequency at each timestep
    #print(len(Pxx))
    dom_freq = []
    for j in range(len(bins)):
        Pxx_transpose = np.transpose(Pxx)
        dom_index = np.argmax(Pxx_transpose[j])
        dom_freq.append(freqs[dom_index])
    #Plot
    axes[row1][col].plot(bins,np.array(dom_freq)/1000)  

    #Label Axes and Colorbar to Scale
    fig.colorbar(f1, ax=axes[row1][col], pad=0).set_label(r'log($\tilde{h}_+(t,f)/M$)', fontsize=14) #add pad=0 term to rrmove spacing
    if col == 0:
        axes[row1][col].set_ylabel('Frequency [kHz]', fontsize=14)
    if row1 == 1:
        axes[row1][col].set_xlabel('Time [s]',fontsize=14)
    
    if Id_f2[k] == True:
        #Make Labels
        label_span = np.linspace(0.004,bins[-1], 50)
        axes[row1][col].plot(label_span, [f2/1000 for i in range(len(label_span))], linestyle='dashed',color = 'darkslateblue')
        axes[row1][col].text(0.003, (f2/1000)-0.1, r'$f_2$', color='darkslateblue' )
    axes[row1][col].text(max(bins)/10,7.25, r'$f_2 \approx$'+f" {round(f2)} Hz", color='darkslateblue')
    axes[row1][col].set_title(f"{id_list2[k]}")


fig.tight_layout()
fig.savefig('./Pictures/Paper_Plots/Spec1.png')
#plt.show()


```
