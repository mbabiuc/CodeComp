Now that we have a table (and Dataframe) summarizing all the tables we want to combine them into one dataframe

```python
import pandas as pd
import numpy as np
```

```python
df_SACRA = pd.read_csv('./Summary_Data/V1_SACRA_Summary.txt', sep='\t')
print(df_SACRA.head())
```

```python
df_SXS = pd.read_csv('./Summary_Data/V1_SXS_Summary.txt', sep='\t')
print(df_SXS.head())
```

```python
df_Whisky = pd.read_csv('./Summary_Data/V1_Whisky_Summary.txt', sep='\t')
print(df_Whisky.head())
```

```python
df_THC = pd.read_csv('./Summary_Data/V1_THC_Summary.txt', sep='\t')
print(df_THC.head())
```

```python
df_BAM = pd.read_csv('./Summary_Data/V1_BAM_Summary.txt', sep='\t')
print(df_BAM.head())
```

```python
list1 = [df_SACRA, df_SXS, df_Whisky, df_THC, df_BAM]
df = pd.concat(list1)
print(df)
```

```python
df.to_csv('./Summary_Data/V1_All_Codes_Summary.txt', sep='\t', index=False)
```
